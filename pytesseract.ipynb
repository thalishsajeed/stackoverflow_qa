```python
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
# function to display an image 
def display_image(filename, length_box=60, width_box=30):
    if type(filename) == np.ndarray:
        image = filename
    else:
        image = cv2.imread(filename)
    plt.figure(figsize=(length_box, width_box))
    plt.imshow(image, cmap="gray")
```

```python
filename = "/home/thalish/dev_data/target.jpg"
```

```python
# No letters detected
print(pytesseract.image_to_string(Image.open(filename)))
```

```python
display_image(filename)
```

```python
# Image cleaning using basic grayscaling and thresholding  
image = cv2.imread(filename)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, threshold = cv2.threshold(gray,55, 255, cv2.THRESH_BINARY)
print(pytesseract.image_to_string(threshold))
```

```python
display_image(threshold)
```

```python
# Image cleaning using basic grayscaling , thresholding and applying gaussian blur 
image = cv2.imread(filename)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, threshold = cv2.threshold(gray,90, 255, cv2.THRESH_BINARY)
blur = cv2.GaussianBlur(threshold, (5, 5), 0)

# To display image on an ipython notebook - ignore when trying from command line
display_image(threshold)
display_image(blur)


print(pytesseract.image_to_string(blur))
```
